```python
# ============================================================
#  MineColab Panel — Celda de inicio
#  1. Ejecuta esta celda
#  2. Abre la URL que aparece abajo en tu navegador
#  3. Gestiona tu servidor desde el panel web
# ============================================================

# Paso A: Montar Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Paso B: Instalar dependencias Python
import subprocess
print("Instalando dependencias...")
subprocess.run([
    'pip', 'install', '-q',
    'flask',
    'flask-socketio',
    'flask-cors',
    'mcrcon',
    'pyngrok',
    'requests',
    'jproperties',
    'PyJWT',
    'psutil',
    'mcstatus',
    'rich'
], check=True)
print("Dependencias instaladas.")

# Paso C: Clonar o actualizar el repositorio
import os

REPO_URL = "https://github.com/Danko3104/Minecraft_Server.git"
PANEL_PATH = "/content/minecolab-panel"

if not os.path.exists(PANEL_PATH):
    print("Clonando repositorio...")
    subprocess.run(['git', 'clone', REPO_URL, PANEL_PATH], check=True)
    print("Repositorio clonado.")
else:
    print("Actualizando repositorio...")
    subprocess.run(['git', '-C', PANEL_PATH, 'reset', '--hard', 'origin/main'], check=True)
    subprocess.run(['git', '-C', PANEL_PATH, 'pull', '--ff-only'], check=True)
    print("Repositorio actualizado.")

# Paso D: Agregar al path e iniciar
import sys
sys.path.insert(0, PANEL_PATH)
os.chdir(PANEL_PATH)

from scripts.start import launch
launch()
```
